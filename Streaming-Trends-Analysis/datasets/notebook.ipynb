```python
import pandas as pd
import matplotlib.pyplot as plt

"""Perform exploratory data analysis 
on the netflix_data.csv data to understand more 
about movies from the 1990s decade.
"""
# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")

"""1. What was the most frequent movie 
duration in the 1990s? Save an approximate answer 
as an integer called duration 
(use 1990 as the decade's start year).
"""
movies_1990s = netflix_df[(netflix_df['release_year'] >= 1990) & (netflix_df['release_year'] < 2000)]
# print(movies_1990s.info())
# print(movies_1990s)
duration = movies_1990s['duration'].mode()[0]
print("Most Frequent Movie Duration (1990s):", duration)

"""2. A movie is considered short 
if it is less than 90 minutes. 
Count the number of short action movies released
in the 1990s and save this integer as short_movie_count.
"""
short_duration_threshold = 90
short_movie_count = movies_1990s[(movies_1990s['genre'].str.contains('Action')) & (movies_1990s['duration'] < short_duration_threshold)].shape[0]

print("Number of Short Action Movies (1990s):", short_movie_count)
```
